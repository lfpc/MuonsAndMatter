```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import moyal

# PDG-inspired constants for iron and muon properties
K = 0.307075         # MeV*cm^2/g
Z = 26               # Iron atomic number
A = 55.85            # g/mol, iron atomic mass
rho = 7.87           # g/cm^3, density of iron
I = 2.86e-4          # Mean excitation energy in MeV (286 eV)
m_e = 0.511          # MeV, electron mass
m_mu = 105.658       # MeV, muon mass
x = 1.0              # cm, thickness of the iron slab
# Radiation length for iron (from PDG)
X0 = 1.76            # cm

# Constant used in the Landau most-probable energy loss formula
j = 0.200

# --- Density Effect Correction ---
def density_effect(beta, gamma):
    """
    Compute the density effect correction (delta) for iron using a Sternheimer 
    parameterization. The parameters below are typical PDG values for iron.
    """
    # Compute x = log10(beta * gamma)
    x_val = np.log10(beta * gamma)
    # Parameters for iron (PDG 2021 typical values)
    x0 = 0.201
    x1 = 3.0
    a = 0.120
    k = 3.0
    C = 3.501
    if x_val < x0:
        return 0.0
    elif x_val >= x1:
        return 4.606 * x_val - C
    else:
        return 4.606 * x_val - C + a * (x1 - x_val)**k

# --- Relativistic Quantities ---
def calculate_relativistic(p):
    """
    Calculate beta and gamma for a muon with momentum p (in MeV/c)
    """
    energy = np.sqrt(p**2 + m_mu**2)
    beta = p / energy
    gamma = energy / m_mu
    return beta, gamma

def T_max(beta, gamma):
    """
    Maximum kinetic energy transferable to an electron in a single collision.
    """
    ratio = m_e / m_mu
    return (2 * m_e * beta**2 * gamma**2) / (1 + 2 * gamma * ratio + ratio**2)

# --- Bethe-Bloch with Density Correction ---
def dEdx_BetheBloch(beta, gamma):
    """
    Calculate the average energy loss per unit length (MeV/cm) using the
    Bethe-Bloch formula with a density effect correction.
    """
    Tmax = T_max(beta, gamma)
    delta = density_effect(beta, gamma)
    # PDG form of the Bethe-Bloch: note the 0.5 factor inside the log.
    term = 0.5 * np.log((2 * m_e * beta**2 * gamma**2 * Tmax) / (I**2)) - beta**2 - delta / 2.0
    dEdx = K * (Z / A) * (1 / beta**2) * term * rho  # MeV/cm
    return dEdx

# --- Landau (Moyal) Parameters ---
def xi_value(beta):
    """
    Calculate the scaling parameter xi (in MeV) used in the Landau distribution.
    """
    return (K / 2) * (Z / A) * (rho * x) / (beta**2)

def calc_delta_p(beta, gamma, xi):
    """
    Calculate the most probable energy loss (calc_delta_p) in MeV.
    """
    arg = (2 * m_e * beta**2 * gamma**2 * xi) / (I**2)
    return xi * (np.log(arg) + j - beta**2)

# --- Multiple Scattering (Molière RMS) ---
def scattering_rms(beta, p):
    """
    Calculate the RMS scattering angle (in radians) using a Molière-type formula.
    p is the initial momentum (in MeV/c).
    
    theta0 = (13.6 MeV)/(beta*p) * sqrt(x/X0) * (1 + 0.038 ln(x/X0))
    """
    factor = 1 + 0.038 * np.log(x / X0)
    theta0 = 13.6 / (beta * p) * np.sqrt(x / X0) * factor
    return theta0

# --- Simulation for One Muon with Correlation Between Loss and Scattering ---
def simulate_muon(p0):
    """
    For a given initial momentum p0 (in MeV/c), simulate one step through iron.
    Returns:
      p0           : initial momentum in MeV/c
      p_final      : final momentum in MeV/c (after energy loss)
      loss         : absolute momentum loss (in MeV)
      theta        : deflection angle in radians (absolute value)
    """
    beta, gamma = calculate_relativistic(p0)
    
    # --- Energy Loss Calculation ---
    avg_dEdx = dEdx_BetheBloch(beta, gamma)   # in MeV/cm
    avg_loss = avg_dEdx * x                    # average energy loss in MeV
    
    # Landau parameters:
    xi_val = xi_value(beta)
    dp_most_prob = calc_delta_p(beta, gamma, xi_val)
    
    # Sample energy loss from a Moyal distribution (approximation to Landau)
    loss_sample = moyal.rvs(loc=dp_most_prob, scale=xi_val)
    # Ensure loss is physically reasonable:
    loss_sample = np.clip(loss_sample, 0, p0)   # cannot exceed initial momentum
    
    # Final momentum after energy loss:
    p_final = p0 - loss_sample
    
    # --- Multiple Scattering ---
    theta0 = scattering_rms(beta, p0)
    # Introduce a correlation: if loss is higher than average, increase scattering slightly.
    # Here we use a simple linear correction factor.
    correlation_factor = 1 + 0.5 * ((loss_sample - avg_loss) / avg_loss)
    effective_theta0 = theta0 * max(correlation_factor, 0.1)  # avoid negative or too small values
    
    # Sample the deflection angle (assumed to be small so Gaussian is a reasonable approximation)
    theta = np.abs(np.random.normal(loc=0.0, scale=effective_theta0))
    # For simplicity, we assume azimuthal symmetry and do not record phi.
    
    return p0, p_final, loss_sample, theta

# --- Simulation Over Many Muons ---
def simulate_many_muons(n_muons=100000):
    """
    Simulate many muons with p0 sampled uniformly (in GeV) between 10 and 350 GeV.
    Returns arrays for:
      p0_all       : initial momenta (in GeV)
      p_final_all  : final momenta (in GeV)
      loss_all     : absolute momentum loss (in GeV)
      theta_all    : scattering deflection angles (in mrad)
    """
    # Sample p0 in GeV and convert to MeV
    p0_GeV = np.random.uniform(10, 200, size=n_muons)
    p0_MeV = p0_GeV * 1000.0  # convert to MeV
    
    p0_all, p_final_all, loss_all, theta_all = [], [], [], []
    
    for p0 in p0_MeV:
        p0_i, p_final_i, loss_i, theta_i = simulate_muon(p0)
        p0_all.append(p0_i)
        p_final_all.append(p_final_i)
        loss_all.append(loss_i)
        theta_all.append(theta_i)
    
    # Convert results to numpy arrays
    p0_all = np.array(p0_all) / 1000.0       # convert to GeV
    p_final_all = np.array(p_final_all) / 1000.0  # GeV
    loss_all = np.array(loss_all) / 1000.0    # GeV
    theta_all = np.array(theta_all) * 1000.0  # convert radians to mrad (1 rad = 1000 mrad)
    
    return p0_all, p_final_all, loss_all, theta_all

# --- Run the Simulation ---
n_muons = 100000
p0_all, p_final_all, loss_all, theta_all = simulate_many_muons(n_muons)

# --- Plotting Histograms ---
plt.figure(figsize=(14, 6))

# Histogram for momentum (energy) loss
plt.subplot(1, 2, 1)
plt.hist(loss_all, bins=100, density=True, alpha=0.7, color='blue')
plt.hist(-np.clip(delta_p*0.01/step_length, -0.1, 0), bins='auto', histtype='step', density=True, label='p0')
plt.hist(delta_p[(step_length<0.011) & (step_length>0.0099)], bins='auto', histtype='step', density=True, label='p0')
plt.xlim(0.01,0.0275)
plt.ylim(0, 350)
plt.xlabel('Momentum Loss (GeV)')
plt.ylabel('Probability Density')
plt.title('Histogram of Muon Momentum Loss')

# Histogram for deflection angle (in mrad)
plt.subplot(1, 2, 2)
plt.hist(theta_all, bins=100, density=True, alpha=0.7, color='red')
plt.xlabel('Deflection Angle (mrad)')
plt.ylabel('Probability Density')
plt.title('Histogram of Muon Deflection Angle')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import levy_stable  # Using stable distribution to approximate Landau

# PDG-inspired constants for iron and muon properties
K = 0.307075         # MeV*cm^2/g
Z = 26               # Iron atomic number
A = 55.85            # g/mol, iron atomic mass
rho = 7.87           # g/cm^3, density of iron
I = 2.86e-4          # Mean excitation energy in MeV (286 eV)
m_e = 0.511          # MeV, electron mass
m_mu = 105.658       # MeV, muon mass
x = 1.0              # cm, thickness of the iron slab
# Radiation length for iron (PDG value)
X0 = 1.76            # cm

# Constant used in the Landau most-probable energy loss formula
j = 0.200

# --- Density Effect Correction ---
def density_effect(beta, gamma):
    """
    Compute the density effect correction (delta) for iron using a Sternheimer 
    parameterization. The parameters below are typical PDG values for iron.
    """
    # Compute x = log10(beta * gamma)
    x_val = np.log10(beta * gamma)
    # Parameters for iron (typical PDG values)
    x0 = 0.201
    x1 = 3.0
    a  = 0.120
    k  = 3.0
    C  = 3.501
    if x_val < x0:
        return 0.0
    elif x_val >= x1:
        return 4.606 * x_val - C
    else:
        return 4.606 * x_val - C + a * (x1 - x_val)**k

# --- Relativistic Quantities ---
def calculate_relativistic(p):
    """
    Calculate beta and gamma for a muon with momentum p (in MeV/c)
    """
    energy = np.sqrt(p**2 + m_mu**2)
    beta = p / energy
    gamma = energy / m_mu
    return beta, gamma

def T_max(beta, gamma):
    """
    Maximum kinetic energy transferable to an electron in a single collision.
    """
    ratio = m_e / m_mu
    return (2 * m_e * beta**2 * gamma**2) / (1 + 2 * gamma * ratio + ratio**2)

# --- Bethe-Bloch with Density Correction ---
def dEdx_BetheBloch(beta, gamma):
    """
    Calculate the average energy loss per unit length (MeV/cm) using the
    Bethe-Bloch formula with density effect correction.
    """
    Tmax = T_max(beta, gamma)
    delta = density_effect(beta, gamma)
    term = 0.5 * np.log((2 * m_e * beta**2 * gamma**2 * Tmax) / (I**2)) - beta**2 - delta/2.0
    dEdx = K * (Z / A) * (1 / beta**2) * term * rho  # MeV/cm
    return dEdx

# --- Landau (Proper) Parameters ---
def xi_value(beta):
    """
    Calculate the scaling parameter xi (in MeV) used in the Landau distribution.
    """
    return (K / 2) * (Z / A) * (rho * x) / (beta**2)

def calc_delta_p(beta, gamma, xi):
    """
    Calculate the most probable energy loss (calc_delta_p) in MeV.
    """
    arg = (2 * m_e * beta**2 * gamma**2 * xi) / (I**2)
    return xi * (np.log(arg) + j - beta**2)

# --- Multiple Scattering (Molière RMS) ---
def scattering_rms(beta, p):
    """
    Calculate the RMS scattering angle (in radians) using a Molière-type formula.
    p is the initial momentum (in MeV/c).
    
    theta0 = (13.6 MeV)/(beta*p) * sqrt(x/X0) * (1 + 0.038 ln(x/X0))
    """
    factor = 1 + 0.038 * np.log(x / X0)
    theta0 = 13.6 / (beta * p) * np.sqrt(x / X0) * factor
    return theta0

# --- Simulation for One Muon with Landau Energy Loss ---
def simulate_muon(p0):
    """
    For a given initial momentum p0 (in MeV/c), simulate one step through iron.
    Returns:
      p0      : initial momentum in MeV/c
      p_final : final momentum in MeV/c (after energy loss)
      loss    : absolute momentum loss (in MeV)
      theta   : deflection angle in radians (absolute value)
    """
    beta, gamma = calculate_relativistic(p0)
    
    # --- Energy Loss Calculation ---
    avg_dEdx = dEdx_BetheBloch(beta, gamma)   # in MeV/cm
    avg_loss = avg_dEdx * x                    # average energy loss in MeV
    
    # Landau parameters:
    xi_val = xi_value(beta)
    dp_most_prob = calc_delta_p(beta, gamma, xi_val)
    
    # Sample energy loss using the Landau distribution.
    # Here we use SciPy’s levy_stable with alpha=1, beta=1 to approximate the Landau.
    loss_sample = levy_stable.rvs(1, 1, loc=dp_most_prob, scale=xi_val)
    # Ensure the loss is physical (non-negative and does not exceed p0)
    loss_sample = np.clip(loss_sample, 0, p0)
    
    # Final momentum after energy loss (note: for highly relativistic muons, energy loss ~ momentum loss)
    p_final = p0 - loss_sample
    
    # --- Multiple Scattering ---
    theta0 = scattering_rms(beta, p0)
    # Introduce a correlation: if loss is higher than average, increase scattering slightly.
    correlation_factor = 1 + 0.5 * ((loss_sample - avg_loss) / avg_loss)
    effective_theta0 = theta0 * max(correlation_factor, 0.1)  # Avoid negative or extremely small values
    
    # Sample the deflection angle (assuming small angles; use Gaussian with sigma = effective_theta0)
    theta = np.abs(np.random.normal(loc=0.0, scale=effective_theta0))
    # (For simplicity we do not track the azimuthal angle)
    
    return p0, p_final, loss_sample, theta

# --- Simulation Over Many Muons ---
def simulate_many_muons(n_muons=100000):
    """
    Simulate many muons with initial momentum p0 sampled uniformly (in GeV) between 10 and 350 GeV.
    Returns arrays for:
      p0_all       : initial momenta (in GeV)
      p_final_all  : final momenta (in GeV)
      loss_all     : absolute momentum loss (in GeV)
      theta_all    : deflection angles (in mrad)
    """
    # Sample p0 in GeV and convert to MeV
    p0_GeV = np.random.uniform(10, 350, size=n_muons)
    p0_MeV = p0_GeV * 1000.0  # convert GeV to MeV
    
    p0_all, p_final_all, loss_all, theta_all = [], [], [], []
    
    for p0 in p0_MeV:
        p0_i, p_final_i, loss_i, theta_i = simulate_muon(p0)
        p0_all.append(p0_i)
        p_final_all.append(p_final_i)
        loss_all.append(loss_i)
        theta_all.append(theta_i)
    
    # Convert to numpy arrays and to proper units:
    p0_all = np.array(p0_all) / 1000.0        # in GeV
    p_final_all = np.array(p_final_all) / 1000.0  # in GeV
    loss_all = np.array(loss_all) / 1000.0      # in GeV
    theta_all = np.array(theta_all) * 1000.0    # convert radians to mrad
    
    return p0_all, p_final_all, loss_all, theta_all

# --- Run the Simulation ---
n_muons = 1000000
p0_all, p_final_all, loss_all, theta_all = simulate_many_muons(n_muons)

# --- Plotting Histograms ---
plt.figure(figsize=(14, 6))

# Histogram for momentum (energy) loss
plt.subplot(1, 2, 1)
#plt.hist(loss_all, bins=100, density=True, alpha=0.7, color='blue')
plt.hist(-np.clip(delta_p*0.01/step_length, -0.1, 0), bins='auto', histtype='step', density=True, label='p0')
plt.hist(delta_p[(step_length<0.011) & (step_length>0.0099)], bins='auto', histtype='step', density=True, label='p0')
#plt.xlim(0.01,0.0275)
#plt.ylim(0, 350)
plt.xlabel('Momentum Loss (GeV)')
plt.ylabel('Probability Density')
plt.title('Histogram of Muon Momentum Loss')

# Histogram for deflection angle (in mrad)
plt.subplot(1, 2, 2)
plt.hist(theta_all, bins=100, density=True, alpha=0.7, color='red')
plt.xlabel('Deflection Angle (mrad)')
plt.ylabel('Probability Density')
plt.title('Histogram of Muon Deflection Angle')

plt.tight_layout()
plt.show()
```
