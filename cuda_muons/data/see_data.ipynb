```python
import pickle
import numpy as np
import gzip
```

```python
with gzip.open('cuda_muons_data.pkl', 'rb') as f:
    data = pickle.load(f)

with open("../plots/hists_a/hists_n2.pkl", "rb") as f:
    hists = pickle.load(f)
hist = hists[(170, 180)]
print(hist.keys())
```

```python
data['py']
```

```python
import matplotlib.pyplot as plt

bin_edges = np.exp(hist["edges_step_length"])
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
plt.bar(bin_centers, hist['hist_step_length'], width=np.diff(bin_edges), align='center', edgecolor='k')
plt.xlabel('exp(step length)')
plt.ylabel('Count')
plt.title('Step Length Histogram')
plt.xscale('log')
plt.show()
```
